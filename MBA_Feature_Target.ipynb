#Juntando tudo

```python
df_historico_abt_train = (
    df_order_items
    .merge(df_orders, on= "order_id", how= "left")
    .query("order_status =='delivered'")
    .query("order_approved_at>='2017-01-01' & order_approved_at	<'2018-07-01'")
    .merge(df_sellers, on= "seller_id", how= "left")
)
```

```python
df_historico_abt_train.shape
```

#Features

```python
df_features_train=(
    df_historico_abt_train
    .query("order_approved_at<'2018-01-01'")
    .groupby("seller_id")
    .agg(uf = ("seller_state","first"),
         tot_ordens_12m = ("order_id","nunique"),
         tot_items_12m = ("product_id","count"),
         tot_items_dist_12m = ("product_id","nunique"),
         receita_12m = ("price","sum"),
         data_ult_vnd=("order_approved_at","max")
         )
    .reset_index()
    .assign(data_ref= pd.to_datetime("2018-01-01 00:00:00"))
    .assign(recencia= lambda df: (df["data_ref"]- df["data_ult_vnd"]).dt.days)
)
```

```python
df_features_train.head()
```

```python
df_features_train.shape
```

#Target

```python
df_target_train=(
    df_historico_abt_train
    .query("order_approved_at >='2018-01-01' & order_approved_at < '2018-07-01'")
    .filter(["seller_id"])
    .drop_duplicates()
)
```

```python
df_target_train.shape
```

```python
df_target_train.head()
```

#Juntando

```python
df_abt_train=(
    df_features_train
    .merge(df_target_train, on= "seller_id", how="left", indicator= True)
    .assign(nao_revendeu_next_6m = lambda df: np.where(df["_merge"]=="left_only",1,0))
    .filter(["data_ref",
             "seller_id",
             "uf",
             "tot_ordens_12m",
             "tot_items_12m",
             "tot_items_dist_12m",
             "receita_12m",
             "recencia",
             "nao_revendeu_next_6m"])
)
```

#Como automatizar o processo de criação de ABT

```python
import dateutil
```

```python
lista_safra = ["2018-01-01", "2018-02-01", "2018-03-01"]

df_abt_safra= pd.DataFrame()

for safra in lista_safra:
  data_ref_safra = pd.to_datetime(safra).date()
  #print(data_ref_safra)
  data_inf_inclusiva = data_ref_safra - dateutil.relativedelta.relativedelta(months = 12)
  dat_sup_exclusiva = data_ref_safra + dateutil.relativedelta.relativedelta(months = 6)
  #print(data_inf_inclusiva)
  #print(dat_sup_exclusiva)

  df_historico_abt_train = (
    df_order_items
    .merge(df_orders, on= "order_id", how= "left")
    .query("order_status =='delivered'")
    .query(f"order_approved_at>='{data_inf_inclusiva}' & order_approved_at	<'{dat_sup_exclusiva}'")
    .merge(df_sellers, on= "seller_id", how= "left")
  )

  df_features_train=(
    df_historico_abt_train
    .query(f"order_approved_at <'{data_ref_safra}'")
    .groupby("seller_id")
    .agg(uf = ("seller_state","first"),
         tot_ordens_12m = ("order_id","nunique"),
         tot_items_12m = ("product_id","count"),
         tot_items_dist_12m = ("product_id","nunique"),
         receita_12m = ("price","sum"),
         data_ult_vnd=("order_approved_at","max")
         )
    .reset_index()
    .assign(data_ref= pd.to_datetime(f"{data_ref_safra}"))
    .assign(recencia= lambda df: (df["data_ref"]- df["data_ult_vnd"]).dt.days)
  )

  df_target_train=(
    df_historico_abt_train
    .query(f"order_approved_at >='{data_ref_safra}' & order_approved_at < '{dat_sup_exclusiva}'")
    .filter(["seller_id"])
    .drop_duplicates()
  )
  df_abt_train=(
    df_features_train
    .merge(df_target_train, on= "seller_id", how="left", indicator= True)
    .assign(nao_revendeu_next_6m = lambda df: np.where(df["_merge"]=="left_only",1,0))
    .filter(["data_ref",
             "seller_id",
             "uf",
             "tot_ordens_12m",
             "tot_items_12m",
             "tot_items_dist_12m",
             "receita_12m",
             "recencia",
             "nao_revendeu_next_6m"])
  )
  df_abt_safra = pd.concat([df_abt_safra, df_abt_train])
```
