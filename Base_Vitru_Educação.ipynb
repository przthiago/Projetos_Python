```python
df = pd.read_excel("Downloads/acompanhamento_de_melhorias.xlsb")
```

```python
df.head()
```

```python
contagem_por_empresa = df[df['Tipo de Melhoria'] == 'Melhoria'].groupby('Empresa')['Tipo de Melhoria'].count()

print(contagem_por_empresa)
```

```python
contagem_tipos_melhoria = df['Tipo de Melhoria'].value_counts()
```

```python
tipo_melhoria_mais_comum = contagem_tipos_melhoria.idxmax()

print("Tipo de Melhoria Mais Comum:", tipo_melhoria_mais_comum)
```

```python
import matplotlib.pyplot as plt
```

```python
contagem_tipos_melhoria = df['Tipo de Melhoria'].value_counts()
```

```python
plt.figure(figsize=(10, 5))
contagem_tipos_melhoria.plot(kind='bar', color='skyblue')
plt.title('Contagem de Melhorias por Tipo')
plt.xlabel('Tipo de Melhoria')
plt.ylabel('Contagem')
plt.xticks(rotation=90)
plt.show()
```

```python
status_counts = df['Status'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Status das Melhorias')
plt.show()
```

```python
# Agrupe os dados por 'Área' e 'Tipo de Melhoria' e conte as ocorrências
area_tipo_counts = df.groupby(['Área', 'Tipo de Melhoria']).size().unstack()
```

```python
# Verifique se há valores nulos e preencha com zero
area_tipo_counts = area_tipo_counts.fillna(0)
```
