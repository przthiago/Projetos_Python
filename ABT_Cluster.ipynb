#ABT Frequencia vs Receita

```python
#junção das bases de df_order_items com df_order com 3 colunas "order_id","seller_id","price" para saber qual a frequencia e receita dos vendedores
abt_fv = (
    df_order_items.merge(df_orders, on="order_id", how="inner")
    .filter(["order_id","seller_id","price"])
    .groupby("seller_id")
    .agg(
        receita=("price","sum"),
        vendas=("order_id", "nunique")
    ).reset_index()
)
```

```python
abt_fv.shape
```

```python
abt_fv.head()
```

```python
#Criando duas colunas rank_pct_receita e rank_pct_vendas para trazer a porcentagem das receita e das vendas
abt_fv["rank_pct_receita"]= abt_fv["receita"].rank(pct=True)
abt_fv["rank_pct_vendas"]= abt_fv["vendas"].rank(pct=True)
```

```python
abt_fv.head()
```

```python
X_abt_fv = abt_fv[["rank_pct_receita","rank_pct_vendas"]]
```

```python
X_abt_fv.head()
```

```python
Kmeans = KMeans(random_state=42)
visualizer= KElbowVisualizer(kmeans, k = (2,10))
visualizer.fit(X_abt_fv)
visualizer.show();
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_abt_fv)
abt_fv["cluster"] = kmeans.labels_
```

```python
abt_fv.head()
```

```python
#Cluster que mais vende é o 2 e o 1 é o que menos vende
sns.scatterplot(data= abt_fv, x= "rank_pct_vendas", y= "rank_pct_receita", hue="cluster", palette="Set1");
```

```python
abt_fv["cluster"].value_counts()
```

```python
#Iremos imprimir o cluster 1, pode passar que esses são os que possívelmente iremos perder
abt_fv.query("cluster==1")
```

```python
#Quer saber a coluna venda é cluste para saber frequencia e receita é a cluster
resultado_final= df.merge(abt_fv, on= "seller_id", how="inner").filter(["seller_id","cluste", "cluster"])
```
