#Crie uma função que indentica o tamanho do terreno

```python
def categorizar(tam):
  if(tam <=500):
    return "pequeno"
  elif (tam >500 and tam < 1000):
    return "medio"
  else: #tam >= 1000
    return "grande"
```

```python
print(categorizar(500))
print(categorizar(900))
print(categorizar(1000))
```

```python
df["categoria_terreno"] = df["tam_terreno"].apply(categorizar)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
%matplotlib inline
df["preco"].plot()
plt.ticklabel_format(scilimits=(-2,8))
```

```python
%matplotlib inline

df["preco"].hist(
    bins=20,
    color="red",
    alpha=0.7,
    grid=False,
    edgecolor="white",
    linewidth=2
)
plt.title("Histograma de Preço")
plt.xlabel("Bins")
plt.ylabel("Quantidade")
plt.ticklabel_format(scilimits=(-2, 7))
```

```python
%matplotlib inline
df[["qtde_quartos","qtde_banheiros"]].hist(
    bins=10,
    color="green",
    alpha=0.7,
    grid=False,
    edgecolor="white",
    linewidth=2
)
```

```python
%matplotlib inline
df.plot(x="qtde_quartos",y="preco", kind="scatter")
plt.ticklabel_format(scilimits=(-2, 7))
```

```python
%matplotlib inline

for col in df.describe().columns:
  if col != "preco":
    df.plot(x=col, y="preco",
        kind="scatter",
        title="{}x preço".format(col),
        color="green")

    plt.ticklabel_format(scilimits=(-2, 7))
```
