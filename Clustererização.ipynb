```python
exemplo= df.assign(total = df.drop("Pais",axis=1).sum(axis=1))
```

```python
exemplo.head()
```

```python
10+	1	+0.5	+8.9	+0.2	+42.3+	0.6+	6+	2
```

```python
for col in exemplo.columns:
  if col not in ["Pais", "total"]:
    #print (col)
    exemplo[col+"_norm"]= exemplo[col]/exemplo["total"]
```

```python
exemplo2=exemplo.filter(regex="_norm").corr()
plt.figure(figsize=(6,6))
sns.heatmap(exemplo2, annot=True, cbar= False, fmt=".2f")
```

```python
df_norm= df_norm.set_index("Pais")
```

```python
df_norm.index
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
label_list= df_norm.index
```

```python
label_list
```

```python
metodo_complete = linkage(df_norm,"complete")
plt.figure(figsize=(15,5))
dendrogram(metodo_complete,labels=label_list)
plt.show()
```

```python
metodo_single = linkage(df_norm,"single")
plt.figure(figsize=(15,5))
dendrogram(metodo_single, labels=label_list)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
```

```python
corte = 0.3
clusters= fcluster(metodo_complete, corte, criterion="distance")
```

```python
clusters
```

```python
df_norm["cluster_id"]= clusters
```
