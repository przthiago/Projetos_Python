##Arima

```python
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
time_series = df.groupby('dt_emissao')['vlr_comis_cli'].sum()
```

```python
plt.figure(figsize=(12, 6))
time_series.plot()
plt.title('Série Temporal de Comissões')
plt.xlabel('Data de Emissão')
plt.ylabel('Valor da Comissão do Cliente')
plt.xticks(rotation=45)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(time_series, model='additive')
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

plt.figure(figsize=(12, 8))

plt.subplot(411)
plt.plot(time_series, label='Original')
plt.legend(loc='upper left')

plt.subplot(412)
plt.plot(trend, label='Tendência')
plt.legend(loc='upper left')

plt.subplot(413)
plt.plot(seasonal, label='Sazonalidade')
plt.legend(loc='upper left')

plt.subplot(414)
plt.plot(residual, label='Resíduo')
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF e PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))
plot_acf(time_series, ax=ax1, lags=40)
plot_pacf(time_series, ax=ax2, lags=40)
plt.show()
```
